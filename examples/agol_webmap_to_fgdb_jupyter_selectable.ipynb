# AGOL Web Map to FGDB Export - Jupyter Version

This notebook lets you: 
- connect to ArcGIS Online / Enterprise
- read a web map
- see all layers in a table
- choose specific layer indexes to export
- export selected layers to a local file geodatabase
- get CSV/TXT/log audit outputs


```python
import pandas as pd
from getpass import getpass
from IPython.display import display

from agol_webmap_to_fgdb_jupyter_selectable import connect_gis, list_webmap_layers, export_selected_layers
```

## 1. Enter your settings

```python
portal_url = 'https://spatialsolutions.maps.arcgis.com/'
webmap_id = '8ed69d6d3dbd4b63ab1a5ec53a081fa2'   # paste web map item id here
username = 'SSS_GISAnalyst1'    
password = 'Temitope2025!'
output_folder = r'C:\Users\OAdeniyi\Downloads\FGDB AGOL Test'
gdb_name = 'webmap_export.gdb'
debug = True

if username and password is None:
    password = getpass('Enter portal password: ')
```

## 2. Connect

```python
gis = connect_gis(portal_url, username=username or None, password=password)
gis
```

## 3. Read the web map and inspect layers

```python
layers_df = list_webmap_layers(gis, webmap_id, include_unsupported=True)
display(layers_df[['index', 'display_title', 'path', 'layerType', 'visibility', 'export_candidate_type', 'supported_for_attempt', 'url']])
```

## 4. Pick the layer indexes you want to export
Example: `[1, 3, 5]`

```python
selected_indexes = [1,2,3,4,5,66,67,68,69,82,83,84]  # example: [1, 2, 4]
selected_indexes
```

## 5. Export selected layers

```python
results_df = export_selected_layers(
    gis=gis,
    webmap_id=webmap_id,
    selected_indexes=selected_indexes,
    output_folder=output_folder,
    gdb_name=gdb_name,
    debug=debug
)
display(results_df)
```
